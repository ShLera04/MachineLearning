В ходе данной работы были  рассмотрены и применены три алгоритма кластеризации: KMeans, DBSCAN и Agglomerative Clustering. Данные были распределены по кластерам с использованием каждого из алгоритмов. Для оценки качества кластеризации использовалась метрика Silhouette Score. 

Silhouette score (силуэтный коэффициент) — это метрика, используемая для оценки качества кластеризации. Она измеряет, насколько объекты внутри кластера похожи друг на друга по сравнению с объектами в других кластерах. Силуэтный коэффициент варьируется от -1 до 1. Значение близкое к 1 указывает на хорошую кластеризацию, где объекты внутри кластера очень похожи друг на друга и сильно отличаются от объектов в других кластерах. Значение близкое к 0 указывает на перекрывающиеся кластеры, где объекты могут быть одинаково близки к объектам в других кластерах. Значение близкое к -1 указывает на плохую кластеризацию, где объекты внутри кластера могут быть более похожи на объекты в других кластерах, чем на объекты внутри своего кластера.

Был проведен анализ распределения целевой переменной y внутри каждого кластера, полученного с помощью алгоритмов. Кластеры с высоким процентом целевой переменной могут быть интересны для дальнейшего анализа, так как они содержат больше объектов с целевым значением 1.

Видим, что у всех трех методов положительный коэффициент, однако значения находятся близко к нулю, в идеальной ситуации оно должно быть близко к 1. Можно сказать, что кластеризация среднего качества. Кластеры не очень хорошо разделены и могут перекрываться.
Agglomerative Clustering и KMeans показывают наилучшие результаты, но общее качество кластеризации среднее. DBSCAN показывает результаты хуже, что может говорить о сложности разделения данных этим методом.
